```python
import pandas as pd

def process_hotel_sales_data(file_path):
    # Read the data
    df = pd.read_csv(file_path)

    # Define the start date
    start_date = pd.to_datetime('01/01', format='%m/%d')

    # Convert the 'Week' column to timedelta and add it to the start date
    df['Date'] = start_date + pd.to_timedelta(df['Week'] - 1, unit='W')

    # Format the 'Date' column to only display the day and month
    df['Date'] = df['Date'].dt.strftime('%d/%m')

    # Select the necessary columns for each year and rename the sales column
    df_2016 = df[['Date', 'Hotel', '2016']].rename(columns={'2016': 'Sales'})
    df_2017 = df[['Date', 'Hotel', '2017']].rename(columns={'2017': 'Sales'})
    df_2018 = df[['Date', 'Hotel', '2018']].rename(columns={'2018': 'Sales'})

    # Add a 'Year' column
    df_2016['Year'] = 2016
    df_2017['Year'] = 2017
    df_2018['Year'] = 2018

    # Concatenate the dataframes
    df = pd.concat([df_2016, df_2017, df_2018])

    # Create a 'Date' column with the format 'day/month/year'
    df['Date'] = df.apply(lambda row: f"{row['Date']}/{row['Year']}", axis=1)

    # Convert the 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
     
    # Drop the 'Year' column
    df = df.drop(columns='Year')

    # Remove comma and $ sign in the 'Sales' column
    df['Sales'] = df['Sales'].str.replace(',', '').str.replace('$', '').astype(int)

    return df
```

```python
df = process_hotel_sales_data('data/Hotels_Sales.csv')
df
```
